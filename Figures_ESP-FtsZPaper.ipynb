```python
import pandas as pd
df_fold = pd.read_csv('TimeSeries-FoldChange.txt', sep = '\t', index_col=0)
df_fold = df_fold.T
```

```python
import matplotlib.pyplot as plt

colors = ['#009999', '#d6a954', '#F47983', '#fa7b58', '#ffea47', '#0698dc', '#326091', '#7abf95']

# Create a figure with 8 subplots (4 rows x 2 columns)
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 10), sharex=True)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each line on its corresponding subplot
for i, (ax, (column, color)) in enumerate(zip(axes, zip(df_fold.columns, colors))):
    ax.plot(df_fold.index, df_fold[column], label=column, color=color)
    #ax.set_title(column)
    ax.legend(loc='best')
    # Add a dotted line at y=0
    ax.axhline(0, color='black', linestyle='--', linewidth=1)

    # Set custom y-axis limits for specific subplots
    if i in [0,1]:
        ax.set_ylim(-5.5, 1.75)
    elif i in [2, 3]:  # Subplots 3 and 4 (zero-indexed)
        ax.set_ylim(-4, 1.5)
    elif i in [4, 5]:  # Subplots 5 and 6 (zero-indexed)
        ax.set_ylim(-3.5, 1.5)
    elif i in [6, 7]:  # Subplots 7 and 8 (zero-indexed)
        ax.set_ylim(-3.5, 2)

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Save the plot as an SVG file
plt.savefig('line_plots_with_custom_y_limits.svg', format='svg')

# Show the plot
plt.show()
```
